```python
import pandas as pd
import numpy as np 
```

```python
df = pd.read_csv('sales_with_nans.csv')
df.head(50)
```

```python
df.isna().sum()
```

```python
total_sales_before = df['Sales_Before'].sum()
print(total_sales_before)
```

```python
total_sales_after = df['Sales_After'].sum()
print(total_sales_after)
```

```python
average_sales_before = df['Sales_Before'].mean()
print(average_sales_before)
```

```python
average_sales_after = df['Sales_After'].mean()
print(average_sales_after)
```

```python
average_satisfaction_before = df['Customer_Satisfaction_Before'].mean()
print(average_satisfaction_before)
```

```python
average_satisfaction_after = df['Customer_Satisfaction_After'].mean()
print(average_satisfaction_after)
```

```python
#Minimum and maximum sales values 
df[['Sales_Before','Sales_After']].max()
```

```python
#Minimum and maximum sales values 
df[['Sales_Before','Sales_After']].min()
```

```python
# Calculate the change in average sales and change in satisfaction from before to after
change_in_sales = average_sales_before - average_sales_after
change_in_sat = average_satisfaction_before - average_satisfaction_after
print(change_in_sales)
print(change_in_sat)
```

```python
# 3️⃣ Filtering & Subsets
# Get all rows for a specific customer_segment e.g high value
# Get all customers with sales above the average
# Filter by multiple conditions (e.g., satisfaction ≥ 4 and sales > 100)
# Drop Rows with missing key values
# fill missing satisfaction values with the mean
# Try filling missing Purchase_Made with "Unknown".
```

```python
# Get all rows for a specific customer_segment e.g high value

high_val_sub = df[df['Customer_Segment'] == 'High Value']
high_val_sub.head(20)
```

```python
# Get all customers with sales above the average

avg_sales = df['Sales_After'].mean()

above_avg_sales = df[df['Sales_After'] > avg_sales]
above_avg_sales.head(20)
```

```python
# Filter by multiple conditions (e.g., satisfaction ≥ 4 and sales > 100)

new_filter = df[(df['Customer_Satisfaction_After'] > 80) & (df['Sales_After'] > 300)]
new_filter.head(20)
```

```python
# Drop Rows with missing key values. In two fields 

df_dropped_col = new_filter.dropna(subset=['Customer_Satisfaction_Before','Sales_Before'])
df_dropped_col.head(20)
```

```python
# fill missing satisfaction values with the mean

filled_vals = df['Customer_Satisfaction_Before'].mean()

filled_vals = df[df['Sales_After'] > avg_sales]
filled_vals.head(20)

filled_vals = df['Sales_After'].mean()
```
