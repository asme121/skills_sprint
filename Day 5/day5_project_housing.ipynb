```python
import pandas as pd 
```

```python
df = pd.read_csv('housing.csv')
df.head(20)
```

```python
import seaborn as sns 
```

```python
#plot a simple scatter 
g = sns.relplot(kind='scatter', data=df, x='longitude', y='latitude', hue='ocean_proximity')
g.set(title='Geographic Location of Houses in California')
```

```python
y = sns.relplot(kind='scatter', data=df, x='longitude', y='latitude', hue='ocean_proximity', size='median_income', palette='YlGnBu')
y.set(title='Geographic Location of Houses in California')
```
